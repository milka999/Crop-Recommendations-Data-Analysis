```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Crop_recommendation.csv")
df = df.dropna()
```

```python
df.head()
```

```python
df.describe()
```

# Which crops can grow in low humidity locales?

```python
plt.figure(figsize=(30,10), dpi= 80)
sns.boxplot(x='label', y='rainfall', data=df, notch=False)
```

The diagram was generated by taking into account data from rainfall datasets.
The range of precipitation for each individual crop type is shown, and it is easy to see which crops
thrive in areas with less than average rainfall (103.479893): chickpeas, beans,
mung beans, lentils, mango, grapes, watermelon and cantaloupe.

# Which crops require a high level og phosphorus to grow? Which ones can thrive in low-phosphorus soil?

```python
plt.figure(figsize=(30,10), dpi= 80)
sns.boxplot(x='label', y='P', data=df, notch=False)
```

The same method was used to determine the amount of phosphorus required for growing different crops. The lowest level
Watermelon and melon require phosphorus, while apples and grapes are crops that require a significantly higher level
phosphorus than others

# Which crops grow at lowest/highest temperatures?

```python
plt.figure(figsize=(30,10), dpi= 80)
sns.barplot(x = 'label',
y = 'temperature',
data = df)
plt.show()
```

The diagram shows the average temperatures at which individual crop types thrive, while
deviation is represented by the black line. Chickpea thrives at the lowest temperature, while papaya thrives
at the highest temperatures.

# Which crops thrive in tropical areas (high temperatures and high humidity)?

```python
plt.figure(figsize=(25,25), dpi=80)
sns.set(style='whitegrid')
sns.scatterplot(x="temperature", y="humidity", hue = 'label', sizes=(30, 200), data = df )
```

The diagram shows all values of the dataset, positioned relative to the values of
temperature and humidity, while different crop types are shown in different colors. We notice that
coconut, papaya and melon cluster in the upper right corner, which means that these are the species that thrive in
areas with the highest temperature and air humidity.

# What Ph level is optimal for growing rice?

```python
plt.figure(figsize=(25,10), dpi=80)
sns.kdeplot(data=df['ph'], label="rice", fill = True)
```

The above histogram was generated based on the ph values of the different soils used for growing rice. We can clearly see that rice thrives best on soil with a pH value of 6 to 7, while the value vary from 3 to 10.5.
